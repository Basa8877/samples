```python
import pandas as pd
import matplotlib.pyplot as plt

# You can replace or add more sequences here
aptamers = {
    "Mango 1": "GGCACGUACGAAGGGACGGUGCGGAGAGGAGAGUACGUGC",
    "Mango 2": "GGCACGUACGAAGGAGAGGAGAGGAAGAGGAGAGUACGUGC",
    }

def calculate_gc(sequence):
    g = sequence.count('G')
    C = sequence.count('C')
    return round(((g+C)/len(sequence))*100,2)

data = []
for name, sequence in aptamers.items():
    gc = calculate_gc(sequence)
    data.append({"aptamer": name, "GC content": gc })
df = pd.DataFrame(data)

plt.figure(figsize=(9,4))
plt.bar(df['aptamer'], df['GC content'], color='purple')
plt.title('GC Content in Aptamer Sequences')
plt.ylabel('GC %')
plt.xlabel('Aptamer')
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
## Aptamer GC Content Analyzer

This notebook calculates the GC content of aptamer sequences using Python and visualizes the results.

**Why GC Content?**
GC-rich regions can affect aptamer folding, stability, and binding affinity. This tool gives a quick view of GC content for diagnostics or design.

Created by: Srivisanth B U
Date: June 2025
```
